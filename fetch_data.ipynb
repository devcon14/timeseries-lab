```python
import requests

def get_glassnode_json(key_name=None, interval="24h"):
    url = "https://api.glassnode.com/v1/metrics/"
    
    if key_name:
        url += key_name
    else:
        url = "indicators/sopr"
        url = "indicators/sopr_adjusted"
        url = "indicators/nvt"
        url = "indicators/stock_to_flow_ratio"
        # age bands
        url = "indicators/soab"
        # stablecoin ratio
        url = "indicators/ssr"
        url = "indicators/puell_multiple"
        url = "addresses/active_count"
        url = "addresses/min_1_count"
        url = "addresses/min_100_count"
        url = "supply/hodl_waves"
        url = "mining/revenue_sum"
        url = "mining/difficulty_latest"
        url = "blockchain/block_interval_mean"
        url = "blockchain/block_count"
        url = "transactions/transfers_volume_to_exchanges_sum"
        url = "transactions/transfers_volume_from_exchanges_sum"
    
    url += "?a=btc"
    # url += "&f=csv"
    url += "&i=" + interval
    print (url)
    url += "&api_key={}".format(API_GLASSNODE)
    
    r = requests.get(url)
    print (r)
    # print (r.content[:100])
    json_data = r.json()
    return json_data

def get_glassnode_df(key_name):
    json_data = get_glassnode_json(key_name)
    print (len(json_data))
    print (json_data[:5])
    import pandas as pd
    df = pd.DataFrame(json_data)
    df['t'] = pd.to_datetime(df['t'], unit='s')
    return df

get_glassnode_json("indicators/net_unrealized_profit_loss")
```
